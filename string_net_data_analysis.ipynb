```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import multiprocessing as mp
import time
import pickle
import os

MY_PATH = "E:/python/pythondata/StringNet/"
FILE_HEADER = "StringNet"
NODE_NAME = "test2"
Num_thread = 4
Num_iteration = 200
v_energy_total = [[] for _ in range(Num_thread)]
c_energy_total = [np.array([]) for _ in range(Num_thread)]
for index in range(Num_thread):
    for iIteration in range(Num_iteration):
        print("index = ",index)
        print("iteration = ", iIteration)
        print('===========================================================')
        file_name = MY_PATH + FILE_HEADER + '_' + NODE_NAME + '_thread_' \
                + str(index) + '_ite_' + str(iIteration)
        
        with open(file_name, 'rb') as File:
            mylist = pickle.load(File)
        v_energy_total[index] = v_energy_total[index] + mylist[1]
        c_energy_total[index] = np.hstack([c_energy_total[index],mylist[0]])
        
```

```python
plt.plot(c_energy_total[0][:200000])
plt.plot(c_energy_total[1][:200000])
plt.plot(c_energy_total[2][:200000])
plt.plot(c_energy_total[3][:200000])

plt.show()
```
