```python
import numpy as np
import matplotlib.pyplot as plt
import bisect
import time
import array


class cLineInfo:
    def __init__(self,y,index,isright):
        self.y = y
        self.index = index
        self.isright = isright
        
class SN:
    def __init__(self,N,M):
        self.N = N
        self.M = M
        self.line_info = [[[cLineInfo(0,i,True),cLineInfo(1,i+N,True)] for _ in range(M)] for i in range(N)]
        self.stack_leg = [[], 2*N]
        self.stack_circle = [[], N]
        self.leg = [0 for i in range(2*N*8)]
        self.cn = N
        self.map = [[i+N for i in range(N)] + [i for i in range(N)] for _ in range(M+1)]
        self.connected_leg = [[i+N for i in range(N)] + [i for i in range(N)] for _ in range(M)]
        self.connected_direction = [[0 for i in range(N)] for _ in range(M)]
        self.circle_index = [i for i in range(N)]
        
        
    # ============  virsualization functions:  =====================
    def fn_virsualization(self):
         
        m = self.N
        for k in range(0,m):
            plt.plot((k,k),(-0.5,self.M+1.5),'y--')     
              
        for my in range(self.M + 1):
            tmap = self.map[my]
            v_finished = [0 for _ in range(2*m)]
            for k in range(0,2*m):
                if v_finished[k]==0:
                    ind1 = k
                    ind2 = tmap[k]
                    if ind1<=m-1:
                        p1_x = ind1
                        p1_y = self.M - my + 1
                    else:
                        p1_x = ind1 - m
                        p1_y = self.M - my              
                    if ind2<=m-1:
                        p2_x = ind2
                        p2_y = self.M - my + 1
                    else:
                        p2_x = ind2 - m
                        p2_y = self.M - my           
                    v_finished[ind1] = 1
                    v_finished[ind2] = 1
                if p1_y==self.M - my and p2_y==self.M - my:
                    self.fn_plot2(p1_x,p1_y,p2_x,p2_y,np.abs(p1_x-p2_x)/m,True) 
                elif p1_y==self.M - my + 1 and p2_y==self.M - my + 1:
                    self.fn_plot2(p1_x,p1_y,p2_x,p2_y,np.abs(p1_x-p2_x)/m,False)
                else:
                    self.fn_plot2(p1_x,p1_y,p2_x,p2_y,np.abs(p1_x-p2_x)/m,True)
        plt.show()
        
    
    """
    def fn_virsualization_leg(self):
        map = self.map
        m = (len(map) - 1)//2     
        for k in range(1,m+1):
            plt.plot((k,k),(-0.5,1.5),'y--')  
        for k in range(len(self.dis1)):
            pipe = self.dis1[k][0]
            y = 1 - self.dis1[k][1]
            if pipe<=self.M-2:
                plt.plot([pipe+1,pipe+2],[y,y],color='r',linewidth=0.2)
            else:
                plt.plot([pipe+1,pipe+1.5],[y,y],color='r',linewidth=0.2)
                plt.plot([0.5,1],[y,y],color='r',linewidth=0.2)
        
        plt.show()
    
    def fn_virsualization_leg2(self):
        
        for k in range(1,self.M+1):
            plt.plot((k,k),(-0.5,1.5),'y--')  
        for x in self.leg:
            # print(x[0])
            if x:
                pipe = x[0]
                y = 1 - x[1]
                if pipe<=self.M-2:
                    plt.plot([pipe+1,pipe+2],[y,y],color='r',linewidth=0.2)
                else:
                    plt.plot([pipe+1,pipe+1.5],[y,y],color='r',linewidth=0.2)
                    plt.plot([0.5,1],[y,y],color='r',linewidth=0.2)
        
        plt.show()
    """
        
        
    def fn_virsualization_leg(self):
        for k in range(0,self.N):
            plt.plot((k,k),(-0.5,self.M + 0.5),'y--')   
        for my in range(self.M):
            for pipe in range(self.N):
                for info in self.line_info[pipe][my]:
                    isright = info.isright        
                    x1 = pipe
                    y1 = - info.y + self.M - my
                    if isright:
                        x2 = x1 + 1
                        y2 = y1
                    else:
                        x2 = x1 - 1
                        y2 = y1
                    if info.index<=2*self.N - 1:
                        plt.plot(x1,y1,'r*')
                    else:
                        plt.plot((x1,x2),(y1,y2),'r*-')
                        print([x1,x2,y1,y2])
                        
        plt.show()
            
                    
                    
                
    
                
    @staticmethod
    def fn_plot2(x1,y1,x2,y2,A,is_up):
        
        if y1!=y2:
            plt.plot((x1,x2),(y1,y2),color='r')     
        elif (is_up):
            vx = np.linspace(x1,x2,500)
            vy = y1 + 0.5*A*np.sin((vx-np.min([x1,x2]))/np.abs(x2-x1)*np.pi)
            plt.plot(vx,vy,color='r')
        else: 
            vx = np.linspace(x1,x2,500)
            vy = y1 - 0.5*A*np.sin((vx-np.min([x1,x2]))/np.abs(x2-x1)*np.pi)
            plt.plot(vx,vy,color='r')
        
        
    def fn_stack_leg_pull(self):
        if not self.stack_leg[0]:
            num_leg = self.stack_leg[1]
            self.stack_leg[1] = num_leg + 1
            # self.leg.append([])
            self.leg.append(0)
            self.leg.append(0)
            self.leg.append(0)
            self.leg.append(0)
            self.leg.append(0)
            self.leg.append(0)
            self.leg.append(0)
            self.leg.append(0)
        else:
            num_leg = self.stack_leg[0][-1]
            self.stack_leg[0].pop()
        return num_leg

    
    def fn_stack_leg_push(self, index):
        self.stack_leg[0].append(index)
        
        
    def fn_stack_circle_pull(self):
        if not self.stack_circle[0]:
            num_circle = self.stack_circle[1]
            self.stack_circle[1] = num_circle + 1
            self.circle.append([])
        else:
            num_circle = self.stack_circle[0][-1]
            self.stack_circle[0].pop()
        return num_circle
    
    
    def fn_stack_circle_push(self,index):
        self.stack_circle[0].append(index)
        
        
    def fn_find_nearest(self,line,m,y): 
        index = 1
        while self.line_info[line][m][index].y < y:
            index = index + 1  
        
        if self.line_info[line][m][index].y == y:
            return None
        else:
            up_ind = self.line_info[line][m][index-1].index
            down_ind = self.line_info[line][m][index].index
            up_isright = self.line_info[line][m][index-1].isright
            down_isright = self.line_info[line][m][index].isright
            
        return up_ind, down_ind, index, up_isright, down_isright
    
    
    def fn_insert_line_info(self,line,m,index,y,index_leg,isright):
        self.line_info[line][m].insert(index,cLineInfo(y,index_leg,isright))
        return True
    
    def fn_insert_line(self,line,m,y,index_leg,isright):
        tc = self.fn_find_nearest(line,m,y)
        index_insert = tc[2]
        print("fn_insert_line, hahahaha")
        print(index_insert)
        if not index_insert == -1:
            self.fn_insert_line_info(line,m,index_insert,y,index_leg,isright)
            
    
    def fn_lieb_add(self, index, p1, p2):
        t1 = self.map[index][p1]
        t2 = self.map[index][p2]
        if not t1 == p2:
            self.map[index][t1] = t2
            self.map[index][t2] = t1
            self.map[index][p1] = p2
            self.map[index][p2] = p1
            
    
    def fn_insert_leg(self,line1,line2,layer,y):
        #line1 = pipe
        #line2 = (pipe + 1) % self.N
        
        map_up = self.map[layer]
        map_down = self.map[layer + 1]
        
        res = self.fn_find_nearest(line1,layer,y)
        if not res:
            return -1
        (left_up,left_down,index1, left_up_isright, left_down_isright) = res
        res = self.fn_find_nearest(line2,layer,y)
        if not res:
            return -1
        (right_up,right_down,index2, right_up_isright, right_down_isright) = res
            
        new_leg = self.fn_stack_leg_pull()
        self.leg[8*new_leg] = left_up
        self.leg[8*new_leg + 2] = left_down
        self.leg[8*new_leg + 4] = right_up
        self.leg[8*new_leg + 6] = right_down
        
        if left_up <= 2*self.N - 1:
            self.leg[8*new_leg + 1] = 0
            self.connected_leg[left_up] = new_leg
            self.connected_direction[left_up] = 4
        else:
            if left_up_isright:
                self.leg[8*new_leg + 1] = 6
                self.leg[8*left_up + 2] = new_leg
                self.leg[8*left_up + 3] = 4
            else:
                self.leg[8*new_leg + 1] = 2
                self.leg[8*left_up + 6] = new_leg
                self.leg[8*left_up + 7] = 4
        
        if left_down <= 2*self.N - 1:
            self.leg[8*new_leg + 3] = 0
            self.connected_leg[left_down] = new_leg
            self.connected_direction[left_down] = 6
        else:
            if left_down_isright:
                self.leg[8*new_leg + 3] = 4
                self.leg[8*left_down + 0] = new_leg
                self.leg[8*left_down + 1] = 6
            else:
                self.leg[8*new_leg + 3] = 0
                self.leg[8*left_down + 4] = new_leg
                self.leg[8*left_down + 5] = 6
                
        if right_up <= 2*self.N - 1:
            self.leg[8*new_leg + 5] = 0
            self.connected_leg[right_up] = new_leg
            self.connected_direction[right_up] = 0
        else:
            if right_up_isright:
                self.leg[8*new_leg + 5] = 6
                self.leg[8*right_up + 2] = new_leg
                self.leg[8*right_up + 3] = 0
            else:
                self.leg[8*new_leg + 5] = 2
                self.leg[8*right_up + 6] = new_leg
                self.leg[8*right_up + 7] = 0
                
        if right_down <= 2*self.N - 1:
            self.leg[8*new_leg + 5] = 0
            self.connected_leg[right_down] = new_leg
            self.connected_direction[right_down] = 2
        else:
            if right_down_isright:
                self.leg[8*new_leg + 7] = 4
                self.leg[8*right_down + 0] = new_leg
                self.leg[8*right_down + 1] = 2
            else:
                self.leg[8*new_leg + 7] = 0
                self.leg[8*right_down + 4] = new_leg
                self.leg[8*right_down + 5] = 2
                
        
        ind = new_leg
        direction = 0
        v_circle1 = []
        while True:
            ind0 = ind
            ind = self.leg[8*ind0 + direction]
            direction = self.leg[8*ind0 + direction + 1]
            if ind<2*self.N-1:
                if ind<self.N-1:
                    ind = ind + self.N
                else:
                    ind = ind - self.N
                (ind_out, v_path_local) = self.fn_combine_leb(map_down,map_up,ind)
                v_circle1 = v_circle1 + v_path_local
                if ind_out<self.N-1:
                    ind_out = ind_out + self.N
                else:
                    ind_out = ind_out - self.N
                ind = self.connected_leg[ind_out]
                direction = self.connected_direction[ind_out]
            if ind == new_leg:
                break
        
        if direction != 0:
            self.cn = self.cn - 1
            if v_circle1:
                v_circle2 = []
                while True:
                    ind0 = ind
                    ind = self.leg[8*ind0 + direction]
                    direction = self.leg[8*ind0 + direction + 1]
                    if ind<2*self.N-1:
                        if ind<self.N-1:
                            ind = ind + self.N
                        else:
                            ind = ind - self.N
                        (ind_out, v_path_local) = self.fn_combine_leb(map_down,map_up,ind)
                        v_circle2 = v_circle2 + v_path_local
                        if ind_out<self.N-1:
                            ind_out = ind_out + self.N
                        else:
                            ind_out = ind_out - self.N
                        ind = self.connected_leg[ind_out]
                        direction = self.connected_direction[ind_out]
                    if ind == new_leg:
                        break
                
                if v_circle2:
                    if len(v_circle1)>=len(v_circle2):
                        self.fn_stack_circle_push(self.circle_index[v_circle2[0]])
                        circle_index = self.circle_index[v_circle1[0]]
                        for change_index in v_circle2:
                            self.circle_index[change_index] = circle_index
                    else:
                        self.fn_stack_circle_push(self.circle_index[v_circle1[0]])
                        circle_index = self.circle_index[v_circle2[0]]
                        for change_index in v_circle1:
                            self.circle_index[change_index] = circle_index
                            
        else:
            self.cn = self.cn + 1
            if v_circle1:
                v_circle2 = []
                direction = 2
                while True:
                    ind0 = ind
                    ind = self.leg[8*ind0 + direction]
                    direction = self.leg[8*ind0 + direction + 1]
                    if ind<2*self.N-1:
                        if ind<self.N-1:
                            ind = ind + self.N
                        else:
                            ind = ind - self.N
                        (ind_out, v_path_local) = self.fn_combine_leb(map_down,map_up,ind)
                        v_circle2 = v_circle2 + v_path_local
                        if ind_out<self.N-1:
                            ind_out = ind_out + self.N
                        else:
                            ind_out = ind_out - self.N
                        ind = self.connected_leg[ind_out]
                        direction = self.connected_direction[ind_out]
                    if ind == new_leg:
                        break
                
                if v_circle2:
                    if len(v_circle1)>=len(v_circle2):
                        circle_index = self.fn_stack_circle_pull()
                        for change_index in v_circle2:
                            self.circle_index[change_index] = circle_index
                    else:
                        circle_index = self.fn_stack_circle_pull()
                        for change_index in v_circle1:
                            self.circle_index[change_index] = circle_index
                            
        self.fn_insert_line_info(line1,layer,index1,y,new_leg,True)
        self.fn_insert_line_info(line2,layer,index2,y,new_leg,False)   
        
        return new_leg
        
                                            
    @staticmethod
    def fn_combine_leb(lieb_up, lieb_down, index0):
        N = len(lieb_up)//2
        if index0 <= N - 1:
            isup = True
        else:
            isup = False
        ind = index0    
        v_mid = []
        
        while True:
            if isup:
                ind = lieb_up[ind]
                if ind <= N - 1:
                    return ind, v_mid
                else:
                    isup = not isup
                    ind = ind - N
                    v_mid.append(ind)
            else:
                ind = lieb_down[ind]
                if ind >= N:
                    return ind, v_mid
                else:
                    isup = not isup
                    v_mid.append(ind)
                    ind = ind + N
            
```
