```python
import numpy as np
import matplotlib.pyplot as plt
import bisect
import time
import array


def fn_lieb_identity(N):
    return [i+N for i in range(N)] + [i for i in range(N)]


"""
def fn_lieb_run(lieb_up, lieb_down, index0, index1, isup0):
        N = len(lieb_up)//2  
        
        isget = False
        ci0 = -1
        
        if isup0:
            ind = index0 + N
        else:
            ind = index0
        isup = isup0
        while True:
            if isup:
                ind = lieb_up[ind]
                if ind<=N-1:
                    if not isget:
                        ci0 = ind
                        isget = True
                    ind = ind + N
                else:
                    ind = ind - N
                    if ind==index0 or ind==index1:
                        return ind, isup, ci0
                isup = not isup
                
            else:
                ind = lieb_down[ind]
                if ind>=N:
                    ind = ind - N
                    if not isget:
                        ci0 = ind
                        isget = True
                else:
                    if ind==index0 or ind==index1:
                        return ind, isup, ci0
                    ind = ind + N
                isup = not isup


def fn_lieb_run_modify(lieb_up, lieb_down, ind0, ind1, isup0, ci_md, v_ci, v_md_ind, v_md_value):
        N = len(lieb_up)//2  
        
        md_num = 0
        
        if isup0:
            ind = ind0 + N
        else:
            ind = ind0
        isup = isup0
        
        while True:
            if isup:
                ind = lieb_up[ind]
                if ind<=N-1:
                    v_md_ind[md_num] = ind
                    v_md_value[md_num] = v_ci[ind]
                    v_ci[ind] = ci_md
                    md_num = md_num + 1
                    ind = ind + N
                else:
                    ind = ind - N
                    if ind==index0 or ind==index1:
                        return md_num
                isup = not isup
                
            else:
                ind = lieb_down[ind]
                if ind>=N:
                    ind = ind - N
                    v_md_ind[md_num] = ind
                    v_md_value[md_num] = v_ci[ind]
                    md_num = md_num + 1
                    v_ci[ind] = ci_md
                else:
                    if ind==index0 or ind==index1:
                        return md_num
                    ind = ind + N
                isup = not isup
""" 

                
class SN:
    def __init__(self,N,M):
        self.N = N
        self.M = M
        self.lmap = [fn_lieb_identity(N) for _ in range(M + 1)]
        self.cn = N
        self.circle_index = [i for i in range(N)]
        self.v_stack = [0 for _ in range(N)]
        self.stack_top = -1
        self.Hamiltonian = [[0 for _ in range(N)] for _ in range(N)]
        
        self.stack_top_backup = -1
        self.cn_backup = N
        self.v_md_stack_ind = []
        self.v_md_stack_val = []
        self.md_stack_num = 0
        self.v_md_map_ind = []
        self.v_md_map_val = []
        self.md_map_num = 0
        self.v_md_circle_ind = []
        self.v_md_circle_val = []
        self.md_circle_num = 0
        
        
    def fn_lieb_add_modify(self, lieb, p1, p2):
        # p1 = p1 + self.N
        # p2 = p2 + self.N
        t1 = lieb[p1]
        t2 = lieb[p2]
        if not t1 == p2:
            lieb[t1] = t2
            lieb[t2] = t1
            lieb[p1] = p2
            lieb[p2] = p1
            self.v_md_map_ind.append(t1)
            self.v_md_map_ind.append(p1)
            self.v_md_map_ind.append(t2)
            self.v_md_map_ind.append(p2)
            self.v_md_map_val.append(p1)
            self.v_md_map_val.append(t1)
            self.v_md_map_val.append(p2)
            self.v_md_map_val.append(t2)
        
        
    def fn_stack_circle_pull(self):
        pop_value = self.v_stack[self.stack_top]
        self.stack_top = self.stack_top - 1
        return pop_value
    
    
    def fn_stack_circle_push(self,index):
        self.stack_top = self.stack_top + 1
        self.v_stack[self.stack_top] = index
        
        
    def fn_lieb_run(self, layer, index0, index1, isup0):
        N = self.N  
        lieb_up = self.lmap[layer]
        lieb_down = self.lmap[layer + 1]
        
        isget = False
        ci0 = -1
        
        if isup0:
            ind = index0 + N
        else:
            ind = index0
        isup = isup0
        while True:
            if isup:
                ind = lieb_up[ind]
                if ind<=N-1:
                    if not isget:
                        ci0 = ind
                        isget = True
                    ind = ind + N
                else:
                    ind = ind - N
                    if ind==index0 or ind==index1:
                        return ind, isup, ci0
                isup = not isup
                
            else:
                ind = lieb_down[ind]
                if ind>=N:
                    ind = ind - N
                    if not isget:
                        ci0 = ind
                        isget = True
                else:
                    if ind==index0 or ind==index1:
                        return ind, isup, ci0
                    ind = ind + N
                isup = not isup

        
        
    def fn_lieb_run_modify(self, layer, ind0, ind1, isup0, ci_md):
        N = self.N 
        
        isget = False
        ci0 = -1
        
        if isup0:
            ind = ind0 + N
        else:
            ind = ind0
        isup = isup0
        
        while True:
            if isup:
                ind = lieb_up[ind]
                if ind<=N-1:
                    if not isget:
                        ci0 = ind
                        isget = True
                    self.v_md_circle_ind.append(ind)
                    self.v_md_circle_val.append(self.v_circle_index[ind])
                    self.v_circle_index[ind] = ci_md
                    md_num = md_num + 1
                    ind = ind + N
                else:
                    ind = ind - N
                    if ind==index0 or ind==index1:
                        return ci0
                isup = not isup
                
            else:
                ind = lieb_down[ind]
                if ind>=N:
                    ind = ind - N
                    if not isget:
                        ci0 = ind
                        isget = True
                    self.v_md_circle_ind.append(ind)
                    self.v_md_circle_val.append(self.v_circle_index[ind])
                    self.v_circle_index[ind] = ci_md
                else:
                    if ind==index0 or ind==index1:
                        return ci0
                    ind = ind + N
                isup = not isup
        
        
    def fn_update_layer_forward(self, layer, legs):
        map_up = self.lmap[layer]
        map_down = self.lmap[layer + 1]
        
        self.cn_backup = self.cn
        self.stack_top_backup = self.stack_top
        
        self.v_md_stack_ind = []
        self.v_md_stack_val = []
        
        self.v_md_map_ind = []
        self.v_md_map_val = []
        
        self.v_md_circle_ind = []
        self.v_md_circle_val = []
        
        for leg in legs:
            p1 = leg[0]
            p2 = leg[1]
            (ind, isup, ci0) = self.fn_lieb_run(layer, p1, p2, True)
            # Case 1:
            if ind==p2 and isup==True:
                self.cn = self.cn + 1
                if ci0!=-1:
                    new_ind = self.fn_stack_circle_pull()
                    ci1 = self.fn_lieb_run_modify(layer,p1,p2,False,new_ind)
                    if ci1!=-1:
                        self.fn_stack_circle_push(new_ind)
                    
            # Case 2:
            # elif ind==p2 and isup==False:
            
            # Case 3:
            elif ind==p1 and isup==False:
                self.cn = self.cn - 1
                if ci0!=-1:
                    ci1 = self.fn_lieb_run_modify(layer,p2,p1,True,ci0)
                    if ci1!=-1:
                        self.v_md_stack_ind.append(self.stack_top + 1)
                        self.v_md_stack_val.append(self.v_stack[self.stack_top+1])
                        self.fn_stack_circle_push(ci1)
                        
            self.fn_lieb_add_modify(self.lmap[layer], p1 + self.N, p2 + self.N)    
                    
                    
```
