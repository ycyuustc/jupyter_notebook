```python
# 通用代码框架

def getHTMLText(url):
    try:
        r = rq.get(url, timeout=30)
        r.raise_for_status() # 如果状态不是200， 引发TRRPError异常
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return "产生异常"


if __name__=="__main__":
    url = "http://www.wipm.ac.cn"
    print(getHTMLText(url))
```
